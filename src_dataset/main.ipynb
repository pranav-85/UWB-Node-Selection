```python
import os
import pandas as pd

directory_path = r"C:\Users\msaip\UWB-Node-Selection\data\UWB raw CIR Dataset1\UWB raw dataset\dataset"

for file_name in os.listdir(directory_path):
    if file_name.endswith(".csv"):
        file_path = os.path.join(directory_path, file_name)
        
        print(f"\nProcessing file: {file_name}")
        
        try:
            df = pd.read_csv(file_path)
            
            if 'LOS' in df.columns:
                counts = df['LOS'].value_counts(dropna=False)
                
                zeros = counts.get(0, 0)
                ones = counts.get(1, 0)
                nulls = df['LOS'].isnull().sum()
                
                print(f"0s: {zeros}, 1s: {ones}, Nulls: {nulls}")
            else:
                print("'LOS' column not found.")
                
        except Exception as e:
            print(f"Error reading {file_name}: {e}")
```
